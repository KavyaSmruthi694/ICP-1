```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score


df = pd.read_csv("glass.csv")

X = df.drop("Type", axis=1)
y = df["Type"]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)

print("\n=== Naive Bayes ===")
print("Accuracy:", accuracy_score(y_test, nb_pred))
print(classification_report(y_test, nb_pred))


svm = SVC(kernel='rbf', C=100, gamma='scale')
svm.fit(X_train_scaled, y_train)
svm_pred = svm.predict(X_test_scaled)

print("\n=== SVM (RBF) ===")
print("Accuracy:", accuracy_score(y_test, svm_pred))
print(classification_report(y_test, svm_pred))


ann = MLPClassifier(hidden_layer_sizes=(50, 30),
                    activation='relu',
                    solver='adam',
                    max_iter=2000,
                    random_state=42)

ann.fit(X_train_scaled, y_train)
ann_pred = ann.predict(X_test_scaled)

print("\n=== ANN (MLPClassifier) ===")
print("Accuracy:", accuracy_score(y_test, ann_pred))
print(classification_report(y_test, ann_pred))

```

```python
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import accuracy_score, classification_report


df = pd.read_csv("glass.csv")

X = df.drop("Type", axis=1)
y = df["Type"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)


nb = GaussianNB()
nb.fit(X_train, y_train)

y_nb = nb.predict(X_test)

print("\n========== NAIVE BAYES ==========")
print("Accuracy:", accuracy_score(y_test, y_nb))
print(classification_report(y_test, y_nb))


svm_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC())
])

svm_params = {
    "svc__kernel": ["rbf"],
    "svc__C": [1, 10, 50, 100],
    "svc__gamma": ["scale", 0.1, 0.01, 0.001]
}

svm_grid = GridSearchCV(
    svm_pipe,
    svm_params,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

svm_grid.fit(X_train, y_train)

print("\nBest SVM Parameters:", svm_grid.best_params_)

svm_best = svm_grid.best_estimator_

y_svm = svm_best.predict(X_test)

print("\n========== SVM (TUNED) ==========")
print("Accuracy:", accuracy_score(y_test, y_svm))
print(classification_report(y_test, y_svm))


ann_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPClassifier(random_state=42))
])

ann_params = {
    "mlp__hidden_layer_sizes": [(50,), (80,40), (100,50)],
    "mlp__alpha": [0.0001, 0.001],
    "mlp__learning_rate_init": [0.001, 0.01],
    "mlp__max_iter": [1500, 3000]
}

ann_grid = GridSearchCV(
    ann_pipe,
    ann_params,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

ann_grid.fit(X_train, y_train)

print("\nBest ANN Parameters:", ann_grid.best_params_)

ann_best = ann_grid.best_estimator_

y_ann = ann_best.predict(X_test)

print("\n========== ANN (TUNED) ==========")
print("Accuracy:", accuracy_score(y_test, y_ann))
print(classification_report(y_test, y_ann))
```

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import accuracy_score, classification_report


df = pd.read_csv("glass.csv")

X = df.drop("Type", axis=1)

y = df["Type"].isin([1, 2]).astype(int)


X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)


svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC(kernel="rbf"))
])

svm.fit(X_train, y_train)

y_svm = svm.predict(X_test)

print("\n========== SVM (Binary, No Tuning) ==========")
print("Accuracy:", accuracy_score(y_test, y_svm))
print(classification_report(y_test, y_svm))


ann = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPClassifier(
        hidden_layer_sizes=(80, 40),
        max_iter=3000,
        random_state=42
    ))
])

ann.fit(X_train, y_train)

y_ann = ann.predict(X_test)

print("\n========== ANN (Binary, No Tuning) ==========")
print("Accuracy:", accuracy_score(y_test, y_ann))
print(classification_report(y_test, y_ann))
```

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import accuracy_score, classification_report


df = pd.read_csv("glass.csv")

X = df.drop("Type", axis=1)

y = df["Type"].isin([1, 2]).astype(int)


X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)


nb = GaussianNB()
nb.fit(X_train, y_train)

y_nb = nb.predict(X_test)

print("\n========== NAIVE BAYES ==========")
print("Accuracy:", accuracy_score(y_test, y_nb))
print(classification_report(y_test, y_nb))


svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC(kernel="rbf"))
])

svm.fit(X_train, y_train)

y_svm = svm.predict(X_test)

print("\n========== SVM ==========")
print("Accuracy:", accuracy_score(y_test, y_svm))
print(classification_report(y_test, y_svm))


single_ann = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPClassifier(
        hidden_layer_sizes=(60,),
        max_iter=3000,
        random_state=42
    ))
])

single_ann.fit(X_train, y_train)

y_single = single_ann.predict(X_test)

print("\n========== ANN (SINGLE LAYER) ==========")
print("Accuracy:", accuracy_score(y_test, y_single))
print(classification_report(y_test, y_single))


multi_ann = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPClassifier(
        hidden_layer_sizes=(80, 40),
        max_iter=3000,
        random_state=42
    ))
])

multi_ann.fit(X_train, y_train)

y_multi = multi_ann.predict(X_test)

print("\n========== ANN (MULTI LAYER) ==========")
print("Accuracy:", accuracy_score(y_test, y_multi))
print(classification_report(y_test, y_multi))
```
