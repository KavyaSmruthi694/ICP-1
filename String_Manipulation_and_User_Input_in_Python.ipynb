```python

def fullname(first_name, last_name):
    return first_name + " " + last_name

def string_alternative(full_name):
    return full_name[::2]

def main():
    first_name = input("Enter first name: ")
    last_name = input("Enter last name: ")

    full_name = fullname(first_name, last_name)
    print("Full name:", full_name)

    print("Output:", string_alternative(full_name))

main()

```

```python
file = open("input.txt", "r")
lines = file.readlines()
file.close()

for line in lines:
    print(line.strip())

print("Word_Count:")

word_count = {}

for line in lines:
    words = line.split()
    for word in words:
        if word in word_count:
            word_count[word] = word_count[word] + 1
        else:
            word_count[word] = 1

out = open("output.txt", "w")

for line in lines:
    out.write(line)

out.write("\nWord_Count:\n")

for word in word_count:
    print(word, ":", word_count[word])
    out.write(word + ": " + str(word_count[word]) + "\n")

out.close()
```

```python
class Employee:

    emp_count = 0

    def __init__(self, name, family, salary, department):
        self.name = name
        self.family = family
        self.salary = salary
        self.department = department

        Employee.emp_count += 1

    def average_salary(self, salaries):
        return sum(salaries) / len(salaries)

    def display(self):
        print("Name:", self.name)
        print("Family:", self.family)
        print("Salary:", self.salary)
        print("Department:", self.department)
        print()


class FulltimeEmployee(Employee):
    def __init__(self, name, family, salary, department):
        super().__init__(name, family, salary, department)



e1 = Employee("Kavya", "Sreeramula", 60000, "HR")

f1 = FulltimeEmployee("Akhil", "Sai", 80000, "IT")
f2 = FulltimeEmployee("Teja", "Sai", 90000, "Finance")

e1.display()
f1.display()
f2.display()

salaries = [60000, 80000, 90000]
print("Average Salary:", e1.average_salary(salaries))

print("Total Employees:", Employee.emp_count)
```
